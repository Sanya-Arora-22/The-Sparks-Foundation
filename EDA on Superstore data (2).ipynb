# Finding the cost

```python
# adding new column 'Cost' in the existing dataset
df['Cost'] = df['Sales'] - df['Profit']
df.head()
     
```

```python
df.groupby("Sub-Category")["Cost"].sum().sort_values(ascending=True).plot(kind="bar")
plt.ylabel("Sum of cost occured")
plt.title("Sub-categorywise sum of cost")
plt.show()
```

Analysis - Retailers spending more cost on Chairs, Phones and Tables

# Finding Profit %

```python
df['Profit %']=(df['Profit']/df['Cost'])*100
#Profit Percentage of first 5 product names
df.iloc[[0,1,2,3,4],[6,]]
```

```python
#Products with high Profit Percentage 
df.sort_values(['Profit %','Sub-Category'],ascending=False).groupby('Profit %').head(20)
     
```
